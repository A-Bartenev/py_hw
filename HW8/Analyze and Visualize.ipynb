# I. Numpy

### Импортируйте NumPy

```python
import numpy as np
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
np_array= np.zeros(10)
print(np_array)
np_array[4]=1
print(np_array)
np_array=np_array.reshape(2,5)
print(np_array)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
np_array= np.arange(10,50)
np_array=np_array[::-1]
print(np_array)
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
np_array=np.random.randint(0, 9,(3, 3))
print(np_array)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
np_array= np.random.rand(4,3,2)
print(np_array)
max_array=np_array.max()
min_array=np_array.min()
print(max_array,min_array)
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
np_array_a= np.random.rand(6,4)
print(np_array_a)
print()
np_array_b= np.random.rand(4,3)
print(np_array_b)
print()
np_array_c=np.matmul(np_array_a,np_array_b)
print(np_array_c)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
np_array= np.random.rand(4,3,2)
print(np_array)
np_std=np_array.std()
print(np_std)
```
